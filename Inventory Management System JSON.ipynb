```python
#Importing the libraries
import json
import time

#Initalizing variables
sale = ''

#Loading the Inventory record from JSON TO String
fd = open('Records.json','r')
js = fd.read()
fd.close()

#Converting String records to JSON/Dictionaries
record = json.loads(js)

#Displaying our Menu
print("--------------------MENU---------------------")
for key in record.keys():
    print(key, record[key]['Name'], record[key]['Price'], record[key]['Qn'])
print("---------------------------------------------")
print('')

#Taking user input about the purcahsing and User Detail
ui_name  = str(input("Enter your Name:"))
ui_mail  = str(input("Enter your E-mail I'd :"))
ui_phone = str(input("Enter your Phone No. :"))
ui_pr    = str(input("Enter product ID : "))
ui_qn    = int(input("Enter Quantity   : "))

print("---------------------------------------------")
print('')

#If we are having eqaul or more quantity that user wants
if (record[ui_pr]['Qn'] >= ui_qn):
    
    #Displaying Purchase Details   
    print("Name        : ", record[ui_pr]["Name"])
    print("Price (Rs)  : ", record[ui_pr]["Price"])
    print("Quantity    : ", ui_qn)
    print("---------------------------------------------")
    bill = ui_qn * record[ui_pr]["Price"]
    print("Billing     : ", bill , "Rs")
    gst_value = bill*(5/100)
    gst = bill* (1 + 5/100)
    print("G.S.T 5%    : ", gst_value , "Rs" )
    print("---------------------------------------------")
    print("Total bill  : ", gst , "Rs")
    #Generating the sales structure in csv format
    sale = ui_name+","+ui_mail+","+ui_phone+","+ui_pr+","+record[ui_pr]['Name']+","+str(ui_qn)+","+ str(record[ui_pr]['Price'])+","+ str(ui_qn*record[ui_pr]['Price'])+","+ time.ctime()+'\n'
    
    #Updating Records Dictionary
    record[ui_pr]['Qn'] = record[ui_pr]['Qn'] - ui_qn
    
#If we are not having enough quantity
else:
    #Display Info about quantity
    print("Sorry, We're not having enough Quantity in our Inventory")
    print("We're only having",record[ui_pr]['Qn'],'quantity')
    print("---------------------------------------------")
    
    #If user want to purchase whole Quantity of the products
    ch = str(input("Press Y/y to purchase"))
    
    #
    if (ch == 'Y' or ch == 'y'):
          
            #Displaying Purchase Details   
            print("Name        : ", record[ui_pr]["Name"])
            print("Price (Rs)  : ", record[ui_pr]["Price"])
            print("Quantity    : ", ui_qn)
            print("---------------------------------------------")
            bill = record[ui_pr]["Price"] * record[ui_pr]['Qn']
            print("Billing     : ", bill, "Rs")
            gst_value = bill*(5/100)
            gst = bill* (1 + 5/100)
            print("G.S.T 5%    :", gst_value )
            print("---------------------------------------------")
            print("Total bill  :", gst)
            #Generating the sales structure in csv format
            sale = ui_name+","+ui_mail+","+ui_phone+","+ui_pr+","+record[ui_pr]['Name']+","+str(record[ui_pr]['Qn'])+","+ str(record[ui_pr]['Price'])+","+ str(record[ui_pr]['Qn']*record[ui_pr]['Price'])+","+ time.ctime()+'\n'
            
            #Updating Records Dictionary
            record[ui_pr]['Qn'] = 0  
    #If user press anything other than Y\y
    else:
           print('Thanks!')

#Converting Dictionary into String
js = json.dumps(record)

#String record in JSON files
fd = open('Records.json','w')
fd.write(js)
fd.close()

#Saving record in sales files
fd = open('Sales.txt', 'a')
fd.write(sale)
fd.close()

print('')
print("---------------------------------------------")
print("  Thanks for your order, Inventory Updated!  ")
print("---------------------------------------------")
```
